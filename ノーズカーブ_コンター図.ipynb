```python
import matplotlib.pyplot as plt
import math 
import numpy as np
import japanize_matplotlib

#定数項
E=2.0



n=1000

z=np.linspace(0, 100, n)
theta=np.linspace(-0.5*math.pi, 0.5*math.pi, n)

Z,Theta=np.meshgrid(z,theta)

r=1
x=1

I=np.zeros((n,n))
V=np.zeros((n,n))
S=np.zeros((n,n))
P=np.zeros((n,n))
for i in range(n):
  for k in range(n):
    I[i][k]=E/((r+Z[i][k]*np.cos(Theta[i][k]))+1j*(x+np.sin(Theta[i][k])))
    V[i][k]=(Z[i][k]*(np.cos(Theta[i][k])+1j*np.sin(Theta[i][k]))*I[i][k])
    S[i][k]=V[i][k]*I[i][k].conjugate()
    P[i][k]=S[i][k].real




plt.contourf(P,abs(V),Theta/math.pi*180)
plt.xlabel("P")
plt.ylabel("V")
plt.colorbar(label="負荷側のインピーダンス角")
plt.show()
```
