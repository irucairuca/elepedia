```python
# -*- coding: utf-8 -*-
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

def main():

    x = np.arange(0, 10.1, 0.5)
    y = np.arange(0, 10.1, 0.5)
    (xm, ym) = np.meshgrid(x, y)        # グリッドの作成
    (xq,yq) = (5, 5)    # 電荷の座標
    r = np.sqrt((xq-xm)**2+(yq-ym)**2)  # 電荷との距離
    k = 9.0*10**9                       # 比例定数k
    q = 1                               # 電荷
    E = (k*q)/r**2  -k*q/(x**2+y**2)                    # 2変数の関数(電場E(x,y)
    (Ey,Ex) = np.gradient(E ,.2, .2)    # Eの勾配を計算
 
    plt.show()


if __name__ == '__main__':
    main()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

coords,ds = np.linspace(-0.6,0.6,100,retstep=True)

(xq1,yq1) = (0.1,0)#q1

(xq2,yq2)=(-0.1,0)#q2



X,Y = np.meshgrid(coords,coords)
k = 9.0*10**9
q1=1
q2=-1

Z=k*q1/((X-xq1)**2+(Y-yq1)**2)+k*q2/((X-xq2)**2+(Y-yq2)**2)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(projection='3d')
#電位のグラフ
ax.plot_surface(X, Y, Z, alpha=0.7)



#勾配の配列は行方向、列方向の順で返されるので注意
#勾配つまり傾きを求める
dY,dX=np.gradient(Z,ds)

Ex=-dX
Ey=-dY
fig,ax=plt.subplots(constrained_layout=True)
#電場のグラフ
plt.streamplot(X,Y,Ex,Ey)

# ax.quiver(X,Y,dX,dY)

```
